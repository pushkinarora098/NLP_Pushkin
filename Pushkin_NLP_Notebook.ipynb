# Train Test Splitting

```python
y=df['cat_id']
X=df.drop(['name','description','featured','shipping_charges','has_options','percentage_discount','applicable_price','product_id','cat_id'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

# Natural Language Processing of column having text

```python
import nltk
import gensim
import re
from bs4 import BeautifulSoup
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
from stop_words import get_stop_words
en=get_stop_words('en')
tokenizer = RegexpTokenizer(r'\w+')
```

```python
cleaned_text=[]
for x in X_train['name_description']:
    text=BeautifulSoup(x).get_text()
    letters_only = re.sub("[^a-zA-Z]", " ", text.lower())
    tokens=tokenizer.tokenize(letters_only)
    stopped_tokens=[i for i in tokens if not i in en]
    stemmed_tokens=[ps.stem(i) for i in stopped_tokens] 
    final_tokens=[i for i in stemmed_tokens if len(i)>=3]
    cleaned_text.append((" ".join(final_tokens)).strip())
```

```python
cleaned_text
```

```python
train_data=pd.DataFrame(cleaned_text, columns=['cleaned_text'])
train_data['cat_id']=y_train
```

```python
train_data.dtypes
```

```python
train_data.head()
```

# converting text to vectors using CountVectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer()
train_vector=cv.fit_transform(train_data['cleaned_text'])
```

```python
train_X=np.array(train_vector.toarray())
train_X
```

```python
test_vector=cv.transform(X_test['name_description'])
test_X=np.array(test_vector.toarray())
```

```python
test_X
```

# Modeling

```python
from sklearn.metrics import f1_score,classification_report,precision_score ,recall_score,accuracy_score 

```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(train_X,y_train)
```

```python
yp=model.predict(test_X)
```

```python
f1_score(y_test,yp,average='weighted')
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,yp)
cm
```
